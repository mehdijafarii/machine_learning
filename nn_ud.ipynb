```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("datasets/fake_reg.csv")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df[["feature1", "feature2"]].values
```

```python
X = df[["feature1", "feature2"]].values
```

```python
y = df["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.max()
```

```python
X_train.min()
```

```python
X_test.max()
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
sqmodel = Sequential(
    [Dense(4, activation="relu"), Dense(2, activation="relu"), Dense(2)]
)
```

```python
from scipy import optimize


sqmodel = Sequential()

sqmodel.add(Dense(4, activation="relu"))
sqmodel.add(Dense(4, activation="relu"))
sqmodel.add(Dense(4, activation="relu"))

sqmodel.add(Dense(1))
#! we can have multiple optimizer and loss function however need to check later
sqmodel.compile(optimizer="rmsprop", loss="mse")
```

```python
sqmodel.fit(x=X_train, y=y_train, epochs=250)
```

```python
sqmodel.history.history["loss"]
```

```python
loss_df = pd.DataFrame(data=sqmodel.history.history["loss"], columns=["loss"])
```

```python
loss_df.head()
```

```python
sns.lineplot(data=loss_df)
```

```python
sqmodel.evaluate(X_test, y_test, verbose=0)
```

```python
sqmodel.evaluate(X_train, y_train, verbose=0)
```

```python
nn_predicts = sqmodel.predict(X_test)
```

```python
nn_predicts.shape
```

```python
pred_vs_actual_df = pd.DataFrame(data=nn_predicts, columns=["Predicted value"])
```

```python
pred_vs_actual_df.head()
```

```python
y_test.shape
```

```python
# changing pandas to numpy array
y_test_array = y_test.values
# changing the shape of the array so now we have 1 coloums and 300 rows
reshape_y_test = y_test_array.reshape(300, 1)
reshape_y_test
```

```python
tem_y_test_df = pd.DataFrame(data=reshape_y_test, columns=["Actual value"])
```

```python
tem_y_test_df.head()
```

```python
pred_vs_actual_df = pd.concat([pred_vs_actual_df, tem_y_test_df], axis=1)
```

```python
pred_vs_actual_df.head()
```

```python
sns.lineplot(data=pred_vs_actual_df["Predicted value"], label="Predicted value")

# Plot the actual values
sns.lineplot(data=pred_vs_actual_df["Actual value"], label="Actual value")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(
    y_pred=pred_vs_actual_df["Predicted value"],
    y_true=pred_vs_actual_df["Actual value"],
)
```

```python
mean_squared_error(
    y_pred=pred_vs_actual_df["Predicted value"],
    y_true=pred_vs_actual_df["Actual value"],
)
```

```python
# ! for root mean square error we need to power it to 0.5
root_mean_squared_error = (
    mean_squared_error(
        y_pred=pred_vs_actual_df["Predicted value"],
        y_true=pred_vs_actual_df["Actual value"],
    )
    ** 0.5
)

root_mean_squared_error
```
