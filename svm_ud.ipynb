```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(data= cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
df['target'] = cancer['target']
```

```python
df.drop('target', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import SVC 
```

```python
X = df
y = cancer['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
init_pred =  model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_pred=init_pred, y_true=y_test))
print(confusion_matrix(y_pred=init_pred, y_true=y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    'C': [1,2,3,111,10000,10,20,30, 0.1],
    'gamma': ['scale', 'auto']
    # 'gamma': [0.1,1, 0.001,0.0001,0.01]
}
```

```python
grid = GridSearchCV(SVC(), param_grid=parameter, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
print(confusion_matrix(y_pred=grid_prediction, y_true=y_test))
print(classification_report(y_pred=grid_prediction, y_true=y_test))
```

```python
print(classification_report(y_pred=init_pred, y_true=y_test))
print(confusion_matrix(y_pred=init_pred, y_true=y_test))
```
